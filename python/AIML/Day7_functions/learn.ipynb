```python
def greet(name) : 
    print(f"Hello! {name}")

greet('Vivek')
```

```python
def greet(name="Bro"):     #function with default parameter
    print(f"Hello, {name}")    

greet()
greet('Arjun')
```

```python
def meet(*name):    #function with flexible parameter
    for n in name :
        print(f"Hello {n} how are you")

meet('Vivek')
meet('Vivek','vishat','amit')
```

```python
#function with return value and params
def add(*vals):
    return sum(vals)

print(add(10,20))

print(add(10,30,50,20))
```
