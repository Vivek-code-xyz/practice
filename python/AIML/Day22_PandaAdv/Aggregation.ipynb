```python

import pandas as pd

# Enhanced student dataset for better groupby examples
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack'],
    'Age': [20, 22, 19, 21, 20, 23, 19, 22, 20, 21],
    'Grade': [85, 92, 78, 88, 95, 72, 89, 81, 90, 87],
    'City': ['New York', 'Boston', 'Chicago', 'Boston', 'New York', 'Chicago', 'Boston', 'New York', 'Chicago', 'Boston'],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math', 'Science', 'Math', 'Science', 'Math', 'Science'],
    'Hours_Studied': [10, 15, 8, 12, 18, 5, 14, 9, 16, 11]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Average grade by city

avg = df.groupby("City")["Grade"].mean()
print(avg)
```

```python
# Total hours studied by city
total_hours_by_city = df.groupby('City')['Hours_Studied'].sum()
print(total_hours_by_city)

# Count students per city
count_by_city = df.groupby('City').size()
print(count_by_city)
```

```python
gradestate = df.groupby("City")["Grade"].agg(["mean","sum","std","max","min","count"])
print(gradestate)
```

```python
#groupby multiple columns
ans = df.groupby(["City","Age"])["Grade"].mean()
print(ans)
```

```python
#diffFun on Diff Cols
answ = df.groupby("City").agg({
    "Grade":"mean",
    "Age":"count",
    "Hours_Studied":"median"
})

print(answ)
```

```python
#custom aggregation
def rangeof(x):
    return x.max()-x.min()

print( df.groupby("City")["Grade"].agg(rangeof) )
```

```python
#filter on aggregate

print( df.groupby("City").filter(lambda x : len(x)>2) )
```

```python
print( df.groupby("City")["Grade"].transform("mean") )
```

```python
df['Diff_From_City_Avg'] = df['Grade'] - df.groupby('City')['Grade'].transform('mean')
print(df)
```
