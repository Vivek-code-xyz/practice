```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace', 'Henry'],
    'Age': [20, 22, 19, 21, 20, 23, 19, 22],
    'Grade': [85, 92, 78, 88, 95, 72, 89, 81],
    'City': ['New York', 'Boston', 'Chicago', 'Boston', 'New York', 'Chicago', 'Boston', 'New York'],
    'Pass': [True, True, True, True, True, False, True, True]
}

df = pd.DataFrame(data)
print(df.to_string())
```

```python
#single column sort
sortbyage = df.sort_values('Grade',ascending=False)

print(sortbyage)
```

```python
MultiSort = df.sort_values(["City","Name"])
print(MultiSort)
```

```python
Multisorted = df.sort_values(["City","Grade"],ascending=[True,False])
print(Multisorted)
```

```python
#sort by index
idx = df.sort_index()
print(idx)

idxs = df.sort_index(ascending=False)
print(idxs)
```

```python
#sort and reset the index as from 0 to n

print( df.sort_values("City").reset_index(drop=True) )
```

```python
#inplace sorting
df1 = df.copy(deep=True)  #creating a copy first

df1.sort_values('Grade',inplace=True)
print(df1)
```
