```python
import pandas as pd


# Creating a student dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace', 'Henry'],
    'Age': [20, 22, 19, 21, 20, 23, 19, 22],
    'Grade': [85, 92, 78, 88, 95, 72, 89, 81],
    'City': ['New York', 'Boston', 'Chicago', 'Boston', 'New York', 'Chicago', 'Boston', 'New York'],
    'Pass': [True, True, True, True, True, False, True, True]
}

df = pd.DataFrame(data)
print(df)
```

```python
toppers = df[df["Grade"]>88]
print(toppers)
```

```python
st_fr_boston = df[df["City"] == 'Boston']
print(st_fr_boston.Name)
```

```python
#multiple conditons
# Students from Boston OR Chicago
print(df[(df["City"] == "Boston")  | (df["City"] == "Chicago")])
```

```python
# Age 20 AND from New York
print( df [ (df["Age"] == 20) & (df["City"]=="New York") ] )
```

```python
# Students NOT from New York
print(df[~(df["City"] == "New York")])
```

```python
# Ages 19 or 22

print( df [df["Age"].isin([19,22])])
```

```python
# Grades between 80 and 90 (inclusive)

num = df[ df["Grade"].between(80,90)]
print(num)
```

```python
#string filters
a = df [ df["Name"].str.startswith("A")]
print(a)

b = df[df["Name"].str.endswith("e")]
print(b)
```

```python
#string cperation

b = df[df["City"].str.contains("or")]
print(b)
```
