# get # update #pop

```python
let_grade = student.get('grade',0) #it returns 0 if grade does not exist
print(let_grade)
```

```python
#update can update or modify many things at once
student.update({"email": "alice@email.com", "age": 21})
print(student)
```

```python
let_st = {"name":"vkek","age":34}
let_st.pop("name")      # Remove and return value
print(let_st.pop("missing", "Not found"))  # Safe removal with default

let_st.popitem()         # Remove and return last item (Python 3.7+)
let_st.clear()           # Remove all items
```
