# Tuples are ordered, immutable sequences created with parentheses ().
# Tuples are IMMUTABLE - you cannot change them

```python
let_tpl = (23,32,54)
print(let_tpl[0])
print(let_tpl[2])
```

```python
# destructuring of tuples
a,b,c = let_tpl
print(b)
```

```python
#creating single tuple 
ab = (23,)      #coma is needed for that
# ab[0]=15        #this is error you cannot modify the existing tuple 
```

```python
#tuple methods

numbers = (1, 2, 2, 3, 2, 4)
print(numbers.count(2))  # 3 (count occurrences)
print(numbers.index(4))  # 3 (first index of value)
```
