```python
#set operations ...add

s = {1,2,3,4}
s.add(5)
s.update([2,3,8,7]) #add multiple elements
print(s)
```

```python
#remove ele

s.remove(3)     # error if not found
s.discard(10)   # no error
s.pop()         # removes random element
s.clear()       # removes all
```

```python
#set operations
s1 = {1,2,3}
s2 = {2,3,4,5}

print(s1|s2) #union ---> merges two sets elements
print(s1&s2)  # intersection  ---> gives common elements
print(s1-s2)   # set difference --> gives element in s1 which is not on s2
print(s1^s2)   #  symetric differece - -> give elements of both the sets which are not common
```
