```python
#read the file

file = open("sample.txt",'r')
content = file.read() 
print(f"file content : {content}")
file.close()
```

Common file modes:
'r'  - Read (default) - file must exist
'w'  - Write - creates new file or OVERWRITES existing
'a'  - Append - adds to end of file
'x'  - Exclusive creation - fails if file exists
'r+' - Read and write
'b'  - Binary mode (e.g., 'rb', 'wb')
't'  - Text mode (default)

```python
#different reading options with creation of file
with open("s.txt",'w') as f :
    f.write("hello everyone \n")
    f.write("how are you all \n i think you were fine \n i am fine too")


```

```python
#reading methods

#1 simple read entire file as one string
with open("s.txt",'r') as file :
    context = file.read()
    print(context)
```

```python
#2 read first n character
with open("s.txt",'r') as file :
    context = file.read(22) 
    print(context)
```

```python
#3 read single line at a time e
with open("s.txt",'r') as file :   #read line  is dynamically iteratable so dont worry just use them wisely
    line1 = file.readline() 
    line2 = file.readline() 
    print(line1,line2,sep=" | ")
```

```python
#4 read multiple lines in a list formate
with open('s.txt','r') as file:
    line_list = file.readlines()
    print(line_list)
```

```python
#5 direct iteration BEST
with open('s.txt','r') as f :
    for lineNo , line in enumerate(f,1) :
        print(f"{lineNo} : {line}")
```

```python
#append mode
with open("s.txt","w") as f:
    f.write("Vivek Khasiya")
```
