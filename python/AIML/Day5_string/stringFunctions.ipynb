now lets functions of Spliting and joining

```python
pystr = "python is great but cpp is best"
words = pystr.split()
print(words)
```

```python
abc="apple,banana,orange,widow"
xyz=abc.split(',')
print(xyz)
```

```python
a="one#two#three#four#five"
b=a.split("#",2)         #it makes only two splits in the string from first to last
c=a.split('#',3)
print(b)
print(c)
```

now comes join functions

```python
wordarr=['you','are','dumb']
sentense = " ".join(wordarr)
sense = "#".join(wordarr)
print(sentense)
print(sense)
```

# usefull other python string functions
text = "Python Programming"             

# Checking content
print(text.startswith('Py'))     # True
print(text.endswith('ing'))      # True
print('gram' in text)            # True
print(text.isalpha())            # False (has space)
print('Python'.isalpha())        # True
print('12345'.isdigit())         # True
print('hello'.islower())         # True

# Finding substrings
print(text.find('Pro'))          # 7 (index where found)
print(text.find('Java'))         # -1 (not found)
print(text.index('Pro'))         # 7 (like find, but raises error if not found)
print(text.count('o'))           # 1 (occurrences)
