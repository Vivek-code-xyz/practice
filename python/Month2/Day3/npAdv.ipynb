```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#meshgrid
a = np.linspace(0,9,10)
a
b = np.linspace(0,9,10)
b
```

```python
xx,yy = np.meshgrid(a,b)
xx
yy
```

```python
plt.scatter(xx,yy)
```

```python
def func(x,y):
    return x**2 + y**2
```

```python
zz = func (xx,yy)
zz
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig  = go.Figure()
fig.add_trace(go.Surface(x=xx,y=yy,z=zz))
fig.show()
```
