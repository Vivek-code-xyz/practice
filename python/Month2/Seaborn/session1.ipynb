```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
#scatter


sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='smoker',size='size')
```

```python
gap = px.data.gapminder()
```

```python
gap
```

```python
gap[gap['country'] == 'India']
```

```python
sns.lineplot(data=gap[gap['country'] == 'India'],x='year',y='lifeExp')
```

```python
sns.relplot(data=gap[gap['country'] == 'India'],x='year',y='lifeExp',kind='line')
```

```python
tempdf = gap[gap['country'].isin(['India','Pakistan','China'])]
```

```python
sns.relplot(data=tempdf,x='year',y='lifeExp',kind='line',hue='country')
```

```python
#facetplot
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',size='size',col='smoker',row='sex')
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='continent',col_wrap=3)
```

```python
#distributive plots
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=10)
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',fill=True)
```

```python
titanic = sns.load_dataset('titanic')

#i want to make the hist plot of male and female survivors of the titanic

sns.displot(data=titanic,x='survived',kind='hist',hue='sex',element='step')
```

```python
#kde plot

sns.displot(data=titanic,x='survived',kind='kde',hue='sex')
```

```python
#rug plot

sns.kdeplot(data=titanic,x='survived',hue='sex')
sns.rugplot(data=titanic,x='survived',hue='sex')
```

```python
sns.histplot(data=tips,x='tip',hue='sex',y='total_bill')
```

```python
#heat map on gap

sns.heatmap(data=gap.pivot(index='year',columns='country',values='lifeExp'))
```

```python
#heatmap on Europian countries 
plt.figure(figsize=(15,15))

sns.heatmap(data=gap[gap['continent'] == 'Europe'].pivot(index='year',columns='country',values='lifeExp'),annot=True,cmap='viridis',linewidths=1)
```
